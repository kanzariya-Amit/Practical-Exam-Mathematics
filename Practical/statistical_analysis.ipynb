# 📊 Statistical Analysis of Student Dataset

Practical Exam Submission

## 📥 Load Dataset

```python
import pandas as pd
import numpy as np

df = pd.read_csv("students_scores.csv")
df.head()
```

## 🔹 Step 1: Measures of Central Tendency & Dispersion

```python
# Mean, Median, Mode
mean_math = df["Math_Score"].mean()
median_math = df["Math_Score"].median()
mode_math = df["Math_Score"].mode()[0]

print("Mean:", mean_math)
print("Median:", median_math)
print("Mode:", mode_math)

# Range, Variance, Standard Deviation
range_science = df["Science_Score"].max() - df["Science_Score"].min()
variance_science = df["Science_Score"].var()
std_science = df["Science_Score"].std()

print("Range:", range_science)
print("Variance:", variance_science)
print("Standard Deviation:", std_science)
```

## 🔹 Step 2: Probability Basics

```python
# Probability of Passing
prob_pass = df["Pass_Fail"].mean()
print("Probability of Passing:", prob_pass)

# Contingency Table
df["Study_Above_5"] = df["Hours_Studied"] > 5
contingency = pd.crosstab(df["Pass_Fail"], df["Study_Above_5"])
print(contingency)

# Conditional Probability
subset = df[df["Hours_Studied"] > 5]
conditional_prob = subset["Pass_Fail"].mean()
print("P(Pass | Hours_Studied > 5):", conditional_prob)
```

## 🔹 Step 3: Distribution & Visualization

```python
import matplotlib.pyplot as plt
from scipy.stats import norm
import numpy as np

# Histogram + Normal Curve
data = df["Math_Score"]
plt.hist(data, density=True)

mean = data.mean()
std = data.std()

x = np.linspace(data.min(), data.max(), 100)
plt.plot(x, norm.pdf(x, mean, std))

plt.title("Histogram + Normal Curve (Math Score)")
plt.show()

# Skewness & Kurtosis
print("Skewness:", df["Science_Score"].skew())
print("Kurtosis:", df["Science_Score"].kurt())

# Q-Q Plot
import scipy.stats as stats
stats.probplot(df["English_Score"], dist="norm", plot=plt)
plt.title("Q-Q Plot (English Score)")
plt.show()
```

## 🔹 Step 4: Linear Algebra Task

```python
# Linear Algebra
math_vec = df["Math_Score"].head(5).values
science_vec = df["Science_Score"].head(5).values

dot_product = np.dot(math_vec, science_vec)

norm1 = np.linalg.norm(math_vec, 1)
norm2 = np.linalg.norm(math_vec, 2)

cos_theta = dot_product / (np.linalg.norm(math_vec) * np.linalg.norm(science_vec))
angle = np.arccos(cos_theta)

print("Dot Product:", dot_product)
print("Norm 1:", norm1)
print("Norm 2:", norm2)
print("Angle (radians):", angle)
```

## 📌 Insights

- Students studying more than 5 hours tend to pass more.
- Math scores approximately follow a normal distribution.
- Science scores may show slight skewness.
- Study hours influence performance significantly.
